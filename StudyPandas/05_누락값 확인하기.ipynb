```python
import numpy as np
import pandas as pd
```

```python
print(np.NaN == True)
print(np.NaN == False)
print(np.NaN == 0)
print(np.NaN == '')
```

```python
print(np.NaN == np.NaN)
print(np.NaN == np.nan)
print(np.NaN == np.NAN)
print(np.nan == np.NAN)
```

```python
print(pd.isnull(np.NaN))
print(pd.isnull(np.nan))
print(pd.isnull(np.NAN))

print(pd.notnull(np.NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

## 에볼라 바이러스

```python
# 에볼라 바이러스 데이터 가져오기
ebola = pd.read_csv('data/country_timeseries.csv')
ebola.head()
```

```python
ebola.info()
```

```python
print(ebola.count())
```

```python
ebola.shape[0] - ebola.count()
```

- 누락값 개수 확인 - count_nonzero(), isnull(), value_counts()

```python
import numpy as np
print(np.count_nonzero(ebola.isnull()))
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
print(ebola['Cases_Guinea'].value_counts(dropna=False))
```

- 누락값 변경하기 - fillna()

```python
ebola.iloc[:, 0:5].head()
```

```python
ebola.fillna(0).iloc[:,0:5].head()
```

```python
ebola.head()
```

```python
ebola['Cases_Guinea'].mean()
```

```python
ebola['Cases_Guinea'].fillna(0).mean()
```

- 누락값을 누락되지 전값으로 채움 : method = 'ffill'

```python
ebola.fillna(method='ffill').iloc[0:5,0:5]
```

```python
ebola['Cases_Guinea'].mean()
```

```python
ebola['Cases_Guinea'].fillna(method='ffill').mean()
```

- 누락값이 나타난 이후 값으로 변경 : method = 'bfill'

```python
ebola.fillna(method='bfill').iloc[0:5,0:5]
```

```python
ebola['Cases_Guinea'].fillna(method='bfill').mean()
```

- 누락값 변경하기 - interpolate()

```python
ebola.interpolate().iloc[0:5,0:5]
```

```python
ebola['Cases_Guinea'].interpolate().mean()
```

```python
ebola.info()
```

```python
ebola.fillna(method='ffill').info()
```

```python
ebola.fillna(method='bfill').info()
```

```python
ebola.interpolate().info()
```

- 누락값 삭제하기 - drop_na()
