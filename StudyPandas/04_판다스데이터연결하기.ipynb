```python
# 열방향으로 데이터 연결하면서 인덱스 정리
print(pd.concat([df1,df2,df3], axis=1, ignore_index=True))
```

- 공통 열과 공통 인덱스 연결하기

```python
df1.columns = ['A','B','C','D']
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']

print(df1)
print(df2)
print(df3)
print(type(df1))
print(type(df2))
print(type(df3))
```

```python
row_concat = pd.concat([df1,df2,df3])
print(row_concat)
```

```python
print(pd.concat([df1,df3]))
print(pd.concat([df1,df3], join='inner'))   # inner : data가 동일하게 가득찬 셀만 합쳐진다(NaN 값 포함 x)
print(pd.concat([df1,df3], join='outer'))   # outer : NaN값 포함
```

```python
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
visited = pd.read_csv('data/survey_visited.csv')
survey = pd.read_csv('data/survey_survey.csv')
```

```python
sv_merge = site.merge(visited, left_on='name', right_on='site')
print(sv_merge)
```

```python
ps_merge = person.merge(survey, left_on='ident', right_on='person')
print(ps_merge)
```

```python
vs_merge = visited.merge(survey, left_on='ident', right_on='taken')
print(vs_merge)
```

- 4개의 데이터 모두 연결
    - survey + person

```python
sp = survey.merge(person, left_on = 'person', right_on='ident')
print(sp)
```

- 4개의 데이터 모두 연결
    - survey + person + visit

```python
spv = visited.merge(sp, left_on='ident', right_on='taken')
spv
```

- 4개의 데이터 모두 연결
    - survey + person + visit + site

```python
spvs = spv.merge(site, left_on='site', right_on = 'name')
spvs
```
