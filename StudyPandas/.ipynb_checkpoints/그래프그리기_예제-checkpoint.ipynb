- 일교차가 가장 큰날

```python
df['일교차'] = df['최고기온(℃)'] - df['최저기온(℃)']
df[df['일교차'] == df['일교차'].max()]
```

- 그래프 표현

```python
import matplotlib.pyplot as plt
plt.plot(df['날짜'][40000:40050], df['평균기온(℃)'][40000:40050])
plt.xticks(df['날짜'][40000:40050], rotation = 90)
plt.show()
```

### 2000년 이후 3월 1일 이후 최고, 최저 기온 확인하기
- 연도/ 월/ 일 분리

```python
df['연도'] = df['날짜'].apply(lambda x : x.split('-')[0])
df['월'] = df['날짜'].apply(lambda x : x.split('-')[1])
df['일'] = df['날짜'].apply(lambda x : x.split('-')[2])
df
```

- 2000년 이후 연도만 리스트로 저장

```python
year = df[ df['연도'] >= '2000']['연도'].unique()
ticks = range(len(year))
print(year)
print(ticks)
```

- 매월 3월 1일의 최고 기온 확인

```python
최고기온 = df[
    (df['연도'] >= '2000')&
    (df['월'] == '03') &
    (df['일'] == '01')
]['최고기온(℃)']
최고기온.index = ticks
print(최고기온)
```

```python
최저기온 = df[
    (df['연도'] >= '2000')&
    (df['월'] == '03') &
    (df['일'] == '01')
]['최저기온(℃)']
최저기온.index = ticks
print(최저기온)
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.plot(최고기온, label='최고기온')
plt.plot(최저기온, label='최저기온')
plt.xticks(ticks, year, rotation=45)
plt.legend()
plt.show()
```

```python
#연도별 평균기온

df_group = df.groupby('연도')['평균기온(℃)'].mean()
print(df_group)
df_group.plot()
```

```python
# 히스토그램으로 표현

plt.hist(df['최고기온(℃)'], bins=100, color='black')
plt.show()
```

## 대중교통 통계자료
### 티머니 : https://www.t-money.co.kr (구글 검색 : 티머니 통계)
- 이용안내 -> 대중교통 통계자료

```python
# 데이터 가져오기

import pandas as pd
df = pd.read_csv('./data/subway_202101.csv', thousands=',')
df.head()
```

```python
# 무임승차 비율 확인

df['무임승차비율'] = df['무임승차'] / (df['무임승차'] + df['유임승차'])
df
```

```python
# 무임승차 비율이 가장 높은 역 전체 현황 확인
df[ df['무임승차비율'] == df['무임승차비율'].max()].iloc[0, 4:8]
```

```python
# 유임승차 비율 = (유임승차 / 전체 승,하차 횟수)*100
station = df[ df['무임승차비율'] == df['무임승차비율'].max() ]
유임승차 = station['유임승차']
유임하차 = station['유임하차']
무임승차 = station['무임승차']
무임하차 = station['무임하차']
total = df[
    df['무임승차비율'] == df['무임승차비율'].max()
].iloc[0,4:8].sum(axis=0)                            #total = station.sum(axis=0)
print(total)
```

```python
# 무임승차 비율이 가장 높은 역의 승,하차 현황 그래프로 표현

import matplotlib.pyplot as aplt
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.pie(
    [
        유임승차.sum() / total * 100, 유임하차.sum() / total * 100,
        무임승차.sum() / total * 100, 무임하차.sum() / total * 100
    ],
    labels=['유임승차','유임하차','무임승차','무임하차'],
    autopct='%.1f%%', startangle = 90)
plt.show()
```

```python
df.columns = columns
df.head()
```

```python
# 역명 + 7~10시 데이터 확인

df2 = df.iloc[:, [3,10,11,12,13,14,15,16]]
df2.head()
```

```python
# 7~10 이동인구 합계

df2['출근'] = df2.iloc[:, 1:].sum(axis=1)
df2.head()
```

```python
# 이동인구 내림차순 정렬
df3 = df2.sort_values(by='출근', ascending=False)
df3.head()
```

```python
# 시간대별 승차인원 합계
mx = df.iloc[:, range(4, 52, 2)].max(axis=0).tolist()
print(mx)
```

```python
# x 축 tick 표시용 '역명(시간)' 형식의 리스트 생성

mx_station = []
for i, e in enumerate(mx):
    if e:
        mx_station.append('%s (%s)'
                         % (df[ df.iloc[:, i*2+4] == e]['지하철역'].iloc[0], i+4))
```

```python
# 시간대별 승차인원이 가장 많은 역 그래프로 표현

import matplotlib.pyplot as plt

plt.bar(range(len(mx_station)), mx, color='blue')
plt.xticks(range(len(mx_station)), mx_station, rotation=90)
plt.show()
```
