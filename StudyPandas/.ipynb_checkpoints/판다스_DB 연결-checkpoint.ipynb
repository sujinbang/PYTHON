## 새로운 연결방법

```python
import sqlalchemy
```

```python
engine = sqlalchemy.create_engine('oracle://scott:tiger@localhost:1521/orcl', encoding='utf-8')
```

```python
conn = engine.connect()
```

```python
pd.read_sql('select * from emp', conn)
```

```python
chunk = pd.read_sql('select * from emp', conn, chunksize=10)
# chunksize : 전체 데이터를 불러올때 너무 큰 용량이라면 사이즈 조절 가능
```

```python
pd.concat([ck for ck in chunk])
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.to_sql('tips',engine, index = False, if_exists='fail')   
# tips 의 raw 개수
#if_exists='replace' : 데이터 소실 위험이 있으므로 최대한 사용하지 않는 걸 추천 - 현 수업에서는 index 열을 삭제하기위해 사용
```
