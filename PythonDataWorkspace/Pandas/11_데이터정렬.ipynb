# 11. 데이터 정렬

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('키') # 키 기준으로 오름차순 정렬
```

```python
df.sort_values('키', ascending=False) # 키 기준으로 내림차순 정렬
```

```python
df.sort_values(['수학','영어'], ascending=False) # 수학, 영어 점수 기준으로 내림차순 정렬
```

```python
df.sort_values(['수학','영어'],ascending=[True, False]) # 수학 점수는 오름차순, 영어점수는 내림차순으로 정렬
```

```python
df['키'].sort_values()
```

```python
df['키'].sort_values(ascending=False)
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```
