# 데이터 선택(조건)

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
```

```python
df[filt]
```

```python
df[-filt]
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학']
```

```python
df.loc[df['키'] >= 185, ['이름','수학','과학']]
```

### 다양한 조건

#### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

#### | 또는

```python
df.loc[(df['키'] < 170) | ( df['키'] > 180)]
```

#### str 함수

```python
filt = df['이름'].str.startswith('송')  # 송씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가는 사람 제외
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)  # SW특기가 Python 이거나 Java인 사람
df[filt]
```

```python
df
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt]
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java',na=True)  # NaN 데이터에 대해서 True로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False)  # NaN 데이터에 대해 False로 설정
df[filt]
```
