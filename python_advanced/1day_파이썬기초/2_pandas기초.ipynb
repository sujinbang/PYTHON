```python
# 원본에 국적명 통합하기 merge()
pd.merge(sample_1, code_master)   #컬럼 개수가 같을때
```

```python
pd.merge(left = sample_1,
        right = code_master,
        how = 'inner',
        left_on = '국적코드',
        right_on = '국적코드')
```

```python
pd.merge(left = sample_1,
        right = code_master,
        how = 'left',           # oracle : left outer join
        left_on = '국적코드',
        right_on = '국적코드')
```

```python
import os
```

```python
path = './files2/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.xlsx')]
```

```python
df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_excel(path + i)
    df = pd.concat([df,data])
    
df = df.reset_index(drop = True)
```
