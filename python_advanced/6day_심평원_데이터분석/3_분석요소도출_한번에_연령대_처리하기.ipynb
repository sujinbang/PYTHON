```python
# files_new 폴더에서 20TABLE 파일 읽어 들인후 조회
# 변수명 : df_20TABLE_org
import pandas as pd
```

```python
file_path = "./files_new2/df_kor_20table_명세서일반내역.xlsx"
df_20TABLE_org = pd.read_excel(file_path)
print(len(df_20TABLE_org))
df_20TABLE_org.head(1)
```

### 월별/연령대별 본인부담비용 시각화분석

```python
# 수진자연령, 요양개시일자, 심결본인부담금
df_month_age = df_20TABLE_org[["수진자연령", "요양개시일자", "심결본인부담금"]]
print(len(df_month_age))
df_month_age.head(1)
```

## 연령대 컬럼 - 한번에 만들기

```python
year_list = ["1대", "10대", "20대", "30대", "40대", "50대", "60대", "70대", "80대", "90대"]

# 모든 연령대를 1개의 데이터프레임으로 통합하기
df_month_age_any = pd.DataFrame()

# 연령대 필터링 범위 계산에 사용
num = 0

for i in range(0, len(year_list), 1) :    
    
    print("연령 필터링 범위 : [수진자연령] >= ", num, " : [수진자연령] < ", num+10)
    
    df_temp =  df_month_age[(df_month_age["수진자연령"] >= num) & (df_month_age["수진자연령"] < num+10)]
    df_temp = df_temp.reset_index(drop=True)
    df_temp["연령대"] = year_list[i]
    df_month_age_any = pd.concat([df_month_age_any, df_temp], ignore_index=True)
    
    num += 10

df_month_age_any = df_month_age_any.reset_index(drop=True)
print()
print("총 행의 갯수 : ", len(df_month_age_any))
df_month_age_any
```
