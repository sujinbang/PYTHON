## 이상치 처리
#### 이상치 처리 이전에 결측치 처리 우선 처리 되어야함 - NA 값은 연산 x

```python
df_temp.describe() # 나이 데이터 이상치 확인
df_temp2 = df_temp.fillna(0)
df_temp2
```

```python
df_temp2[df_temp['나이'] >= 100]
```

```python
# df_temp[['나이']].boxplot()
plt.boxplot(df_temp2['나이'])
```

- 최대극단치: Q3 + 1.5 * iqr 
- 최소극단치 : Q1 - 1.5 * iqr

```python
# 25,75 사분위수 확인 : np.percentile()
q1,q3 = np.percentile(df_temp2['나이'],[25,75])
print(q1)
print(q3)
```

```python
iqr = q3-q1
iqr
```

```python
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
print(upper_bound)
print(lower_bound)
```

```python
condition = (df_temp2['나이'] > upper_bound) | (df_temp2['나이'] < lower_bound)
df_temp2[condition]
```

```python
def x (dframe, data):
    q1,q3 = np.percentile(data,[25,75])
    iqr = q3 - q1
    
    upper_bound = q3 + (1.5 * iqr)
    lower_bound = q1 - (1.5 * iqr)
    
    print(upper_bound)
    print(lower_bound)

    condition = (data > upper_bound) | (data < lower_bound)
    print(len(dframe[condition]))
    
    return dframe[condition]
```

```python
df_temp2['이용거리']
```

```python
temp = x(df_temp2, df_temp2['이용거리'])
temp
```
