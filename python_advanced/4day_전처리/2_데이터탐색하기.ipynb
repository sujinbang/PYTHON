#### EDA : 탐색적데이터 처리

```python
import pandas as pd
```

```python
file_path = './data/exam_sample.csv'
df = pd.read_csv(file_path)
df
```

```python
df.describe()
```

```python
#class 데이터를 기준으로 그룹을 지은 후에 df1 변수에 넣어주세요
df1 = df.groupby(['class'])
```

```python
df1
```

```python
df1.head(1)  #  그룹 당 첫번째값 출력
```

```python
df1.head()
```

```python
# 특정 그룹 조회
df1.get_group('C')
```

```python
# class 그룹별 각 과목에 대한 평균

df.groupby('class').mean()

# 숫자형 데이터의 칼럼만 평균 계산하여 출력
```

```python
# 성별로 그룹
df.groupby('sex').head()
```

```python
# 반별 및 성별 그룹화 하기
df_temp = df.groupby(['class','sex'])
df_temp.head(1)
```

```python
# A반의 남성에 대해서만 조회해 주세요
df_temp.get_group(('A','m'))
```

```python
# 성별 각 과목의 평균 확인하기
df.groupby('sex').mean()
```

```python
# 반별 성별 각 과목 평균
df.groupby(['class','sex']).mean()
```

```python
# 반별 수학점수의 평균 조회
df.groupby('class')['math'].mean()
```

```python
df['math'].groupby(df['class']).mean()
```

```python
df.groupby('class').mean()[['math']]  # [[]] : 데이터 프레임 형태로 출력
```

```python
# 반별로 수학 응시생 수를 조회해 주세요
df['math'].groupby(df['class']).count()
```

```python
df.groupby('class')['math'].count()
```
