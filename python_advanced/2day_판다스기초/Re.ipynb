```python
# sample 파일 총 24개를 읽어들어서
# sample_total 데이터프레임에 합쳐서 넣을겁니다
# 최종 컬럼 : 국적코드, 성별, 입국객수, 기준년월, 국적명
# sample_total 조회
# 기준년월 : 각 파일당 2019년 1월 ~ 2020년 12월 값 사용
```

```python
# 파일 개수 - 24
# 국적명
# 기준년월 - 24
```

```python
import pandas as pd
```

```python
# sample 파일 24개 불러오기

total = pd.DataFrame() #빈 데이터프레임 하나를 생성한다

for i in range (0, 24):
    file_name = "files2/sample({}).xlsx".format(i)
    temp = pd.read_excel(file_name, header=1, skipfooter=2, usecols=[0,1,2])
    total = pd.concat([total, temp]) #전체 데이터프레임에 추가하여 넣는다

print(len(total))
total.head(10)
```

```python
cntry = pd.read_excel('./files2/sample_codemaster.xlsx')
cntry
```

```python
total = pd.merge(total,
         cntry,
         on = '국적코드',
         how = 'left')
total
```
