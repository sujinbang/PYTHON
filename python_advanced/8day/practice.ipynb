```python
import pandas as pd

df_total = pd.DataFrame()

for i in range(1,81,1):
    file_path = './포항/trfcard({})/trfcard.csv'.format(i)
    df = pd.read_csv(file_path)
    df_total = pd.concat([df_total,df])
    
```

```python
file_path = './포항/trfcard(1)/trfcard_columns.xlsx'
col_kor = pd.read_excel(file_path, header = 2)
col_kor.iloc[:,[1,2]]
```

```python
col_kor_list = []
for j in range(len(col_kor)):
    col_kor_list.append(col_kor['컬럼명 (한글)'][j])
                        
print(col_kor_list)

df_total.columns = col_kor_list
```

```python
df_total['체류시간'] = pd.to_datetime(df_total['하차시각']) - pd.to_datetime(df_total['승차시각'])
```
