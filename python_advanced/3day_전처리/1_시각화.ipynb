```python
# 피벗테이블 이용하여 상위 top5 조회
# 국적별 전체비율에 대한 평균을 내림차순하여 상위 5개 추출
temp = df.pivot_table(values = '전체비율(%)',
              index = '국적',
              aggfunc = 'mean').sort_values('전체비율(%)',
                                           ascending = False).head()
```

```python
temp
```

```python
# 데이터 프레임 형태로 변환하기
temp.reset_index(inplace = True)
```

```python
temp
```

```python
cntry_list = temp['국적']
cntry_list
```

```python
for cntry in cntry_list:           #for문의 값을 이용한 반복문
    
    # 중국 데이터만 조회하기
    condition = df['국적'] == '{}'.format(cntry)
    condition
    
    df_filter = df[condition]
    df_filter

    # 그래프 크기 조정
    plt.figure(figsize = (12,4))    #(너비, 높이)
    # 데이터 넣기
    plt.plot(df_filter['기준년월'],df_filter['관광'])
    
    # 그래프 전체 제목 넣기
    plt.title(f'{cntry} 국적의 년월별 관광객 추이')

    # x축, y축 이름 달기
    plt.xlabel("기준년월")
    plt.ylabel("관광객수")

    # x축의 눈금 값 설정하기
    plt.xticks(['2010-01','2011-01','2012-01','2013-01','2014-01'
                ,'2015-01','2016-01','2017-01','2018-01','2019-01'])
    


    plt.show()
```

```python
# 중국
# 년도 및 월별 관광 데이터 조회
# 행은 년도, 열은 월
df_pivot = df_filter.pivot_table(values = '관광',
                          index = '년도',
                          columns = '월')
df_pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 그래프 사이즈 설정하기
plt.figure(figsize = (16,10))

#히트맵 그리기
sns.heatmap(df_pivot, annot=True, fmt='.0f', cmap='rocket_r')

plt.title('년도 및 월별 중국 관광객 추이(히트맵)')
plt.show()
```

```python
for cntry in cntry_list:
    condition = df['국적'] == f'{cntry}'
    df_filter = df[condition]

    # 중국
    # 년도 및 월별 관광 데이터 조회
    # 행은 년도, 열은 월
    df_pivot = df_filter.pivot_table(values = '관광',
                              index = '년도',
                              columns = '월')

    # 그래프 사이즈 설정하기
    plt.figure(figsize = (16,10))

    #히트맵 그리기
    sns.heatmap(df_pivot, annot=True, fmt='.0f', cmap='rocket_r')

    plt.title('년도 및 월별 {} 관광객 추이(히트맵)'.format(cntry))
    plt.show()
```
